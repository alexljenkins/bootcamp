# Altair, Plotly and D3 (built on Java)

```python
import matplotlib.pyplot as plt
```

```python
#create a loop to create the x and y values
x=[]
y=[]
for i in range(-10,11,1):
    x.append(i)
    y.append(i**2)
```

```python
plt.figure(figsize=(10,7)) #creates the canvas (not required but if you want to change the size)
plt.plot(x,y) #plots the graph
plt.title('Polynomial $y = x^2$') #adds title etc
```

# Pandas

```python
import pandas as pd
```

```python
#to plot with pandas, must be in dataframe. But this also allows multiple graphs to be plotted together from a single df
df = pd.DataFrame(index = x, data=y, columns=['y'])
```

```python
#Super simple to plot with, if standard line graph
#df.plot()
#plt.axis([-15,15,0,150])

#This draws a scatter plot. Need x and y as non-index values
df = pd.DataFrame(data={'x':x,'y':y})
df.plot('x','y', kind='scatter')
#after plotting with pandas we can now use matplotlib to design the graph
```
